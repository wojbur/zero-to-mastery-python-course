```python
import seaborn as sns
sns.set()

graph = sns.scatterplot(x='wage_eur', y='value_eur', data=df1)
```

```python
from bokeh.plotting import figure, show
from bokeh.models import HoverTool

TOOLTIPS = HoverTool(tooltips=[
    ("index", "$index"),
    ("(wage_eur,value_eur)", "(@wage_eur, @value_eur)"),
    ("short_name", "@short_name"),
])

p = figure(title="FIFA22", x_axis_label='wage_eur', y_axis_label='value_eur', plot_width=700, plot_height=700, tools=[TOOLTIPS])
p.circle('wage_eur', 'value_eur', size=10, source=df1)
show(p)
```
